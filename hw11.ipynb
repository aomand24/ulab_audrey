**Orbital Periods and Kepler's Third Law** \
Kepler's Third Law states that the square of the orbital period $T$ (in years) of a planet is proportional to the cube of the semi-major axis $a$ (in astronomical units, AU):

$$
T^2 = a^3
$$

```python
def orbital_period(a):
    period= (a**3)**0.5
    return period
```

You are provided with a list of exoplanet radii (in Earth radii) and their equilibrium temperatures (in Kelvin):

`radii = [1.2, 0.8, 1.5, 2.0, 1.0, 0.6]` \
`temperatures = [300, 400, 350, 280, 500, 450]`

Write a loop to filter and print the radii of planets with temperatures below 350 K.

```python
radii = [1.2, 0.8, 1.5, 2.0, 1.0, 0.6]
temperatures = [300, 400, 350, 280, 500, 450]

def planets_under_350(temperatures, radii):
    #filters the radii and only gives a list of radii who have a corresponding temperature under 350 K that is not inclusive
    filtered_radii=[]
    for temp in range(len(temperatures)):
        if temperatures[temp] < 350:
            filtered_radii.append(radii[temp])
    return filtered_radii

planets_under_350(temperatures, radii)
```

```python
mean=np.mean(array)
median=np.median(array)

def mode(list):
    #finds the number that is most frequent in the whole array
    freq={}
    for i in list:
        freq.setdefault(i,0)
        freq[i] += 1
    hf=max(freq.values())
    hflst=[]
    for i,j in freq.items():
        if j == hf:
            hflst.append(i)
    return hflst
    
print(mean,median,mode(array))
```
