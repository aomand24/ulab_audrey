```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def five_sine_functions():
# takes a range of A values and w values and runs them through a given function A*sin(w*x) and plots 5 functions.
    A_values=[1,2,3,4,5]
    w_values=[1,2,3,4,5]
    x=np.linspace(0,2*np.pi,1000)
    plt.figure(figsize=(10,6))
    for A,w in zip(A_values,w_values):
        y=A*np.sin(w*x)
        plt.plot(x,y,label=f"A={A},w={w}")
    plt.xlabel("x data")
    plt.ylabel("y data")
    plt.title("Five Sine functions")
    plt.legend()





five_sine_functions()
```

```python
import random
list_1=[random.randint(0,100) for _ in range(40)]
list_2=[random.randint(0,100) for _ in range(40)]
x_vals_1=list(range(len(list_1)))
x_vals_2=list(range(len(list_2)))
plt.figure(figsize=(8,10))
plt.plot(x_vals_1,list_1,linewidth=10,label=f"index,list_1={list_1}")
plt.plot(x_vals_2,list_2,color="red",linestyle="--",label=f"index,list_2={list_2}")
plt.xlabel("index")
plt.ylabel("value")
plt.title("two lines")
plt.legend()

plt.show()
```
