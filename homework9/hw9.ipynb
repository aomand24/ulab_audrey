# 2 Plotting a Function
Goal: Write a **function** that will plot $f(x) = x^2$ and $g(x) = \sqrt{x}$ on the same plot and return the $f(x)$ and $g(x)$ values as separate lists. Your plot needs to have the following:
- An appropriate figure size (`figsize=(10,8)` is recommended)
- Labeled x and y axis
- A title
- Different colors and linestyles for each function
- A legend
- Your domain should be $x \in [0, 10]$. *Hint: Use `np.linspace` to set a range of values for your functions.*

Your function should return:
- A list of $f(x)$ values
- A list of $g(x)$ values.

```python
def plot_and_return(x_range):
# takes a range of x values and plots two functions, x squared and square root x on the same graph and returns two lists with the values of each function.
    x_values=np.linspace(x_range[0],x_range[1],5)
    y_vals_f=(x_values)**2
    y_vals_g=np.sqrt(x_values)
    plt.plot(x_values, y_vals_f, label='x squared', color='b')
    plt.plot(x_values, y_vals_g, label='square root x', color='r')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Plot of Two Functions')
    plt.legend()
    plt.show()
    return print("f(x):",list(y_vals_f)), print("g(x):",list(y_vals_g))

plot_and_return((0,10))
```

```python
import numpy as np
import matplotlib.pyplot as plt
def plot_in_3d():
# takes range of x values and y values and creates a meshgrid that is plotted in 3d
    x=np.linspace(0,5,100)
    y=np.linspace(0,5,100)
    X, Y = np.meshgrid(x, y)
    Z = np.cos(np.sqrt(X**2 + Y**2))
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    surface = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    ax.set_title('3D Meshgrid Plot')
    fig.colorbar(surface)
    plt.show()

plot_in_3d()
```
